```python
import re
import glob
import pandas as pd
from collections import defaultdict
```

```python
# Have to be adjusted according to your env
path_regexp = "../../jmartin/work/trio1.offspring.*/MultiQC/*/multiqc_hicpro.txt"
```

```python
multiqc_hicpro_files = glob.glob(path_regexp)
multiqc_hicpro_files
```

#### Concatenate all multiqc_hicpro files

```python
protocols = []
for multiqc_hicpro in multiqc_hicpro_files:
    protocol =  re.search("trio1.offspring.(?P<protocol>[\w\-]+)", multiqc_hicpro).group("protocol")
    df = pd.read_csv(multiqc_hicpro, sep="\t") 
    df['protocol'] = protocol
    protocols.append(df)
```

```python
df = pd.concat(protocols, ignore_index=True)
columns = df.columns.to_list()
df
```

#### Keeping only a subset of informations

```python
to_keep = ['Sample','Reported_pairs', 'Valid_interaction_pairs', 'percent_paired_reads', 'Dangling_end_pairs', 'protocol']
columns_to_drop = list(set(columns) - set(to_keep))
df.drop(labels=columns_to_drop, axis=1, inplace=True)
df
```

```python
df.drop(labels=['Sample'], axis=1, inplace=True)
df
```

#### Aggregate within each protocol

```python
aggregation_functions = defaultdict()
for column in df.columns:
    if column == "protocol":
        aggregation_functions[column] = 'first'
    elif "percent" in column:
        aggregation_functions[column] = 'mean'
    else:
        aggregation_functions[column] = 'sum'
```

```python
df_new = df.groupby(['protocol']).aggregate(aggregation_functions)
df_new.drop(labels=['protocol'], axis=1, inplace=True)
display(df_new)
```
